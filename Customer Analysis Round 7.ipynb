```python
def one_hot_encoder(df):
    df_list = []
    for column in df.columns:
        encoder = OneHotEncoder()
        encoder.fit(df[[column]])
        encoded = encoder.transform(df[[column]]).toarray()
        df_list.append(pd.DataFrame(encoded,columns=encoder.categories_))
    new_df = pd.concat(df_list, axis=1)
    return new_df
        
        
```

```python
data_categorical_hot = data_categorical_cleaned.loc[:,["Response", "Gender", "Sales Channel"]]
```
